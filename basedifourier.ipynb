```python
import numpy as np
import matplotlib.pyplot as plt

# Parametri
N = 8  # dimensione dello spazio
n = np.arange(N)  # indici temporali

# Funzione per generare il k-esimo vettore di Fourier
def fourier_basis_vector(k, N):
    return (1 / np.sqrt(N)) * np.exp(1j * 2 * np.pi * k * n / N)

# Plot di alcuni vettori
ks = [0, 1, 2, N//2]  # valori di k da visualizzare
plt.figure(figsize=(20, 12))

for i, k in enumerate(ks, 1):
    f_k = fourier_basis_vector(k, N)
    
    # Parte reale
    plt.subplot(len(ks), 2, 2*i - 1)
    plt.stem(n, np.real(f_k), basefmt=" ")
    plt.title(f"Parte reale - k = {k}")
    plt.xlabel("n")
    plt.ylabel("Re[f^(k)]")
    plt.grid(True)
    
    # Parte immaginaria
    plt.subplot(len(ks), 2, 2*i)
    plt.stem(n, np.imag(f_k), basefmt=" ")
    plt.title(f"Parte immaginaria - k = {k}")
    plt.xlabel("n")
    plt.ylabel("Im[f^(k)]")
    plt.grid(True)

plt.tight_layout()
plt.show()
```
